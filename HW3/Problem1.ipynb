**Task:** to use the BFS algorithm to return the kth-level friend for a given user and graph: FindFriends(G, User, K)

```python
from collections import deque

def FindFriends(g, user, k):
    # check if user is in graph
    if user not in g:
        print(f"User {user} not found in the graph.")
        return

    # initialize visited queue and current queue
    visited = set()
    queue = deque([(user, 0)])

    while queue:
        # pop the current user and its level
        current_user, level = queue.popleft()

        # check if current level matches target
        if level == k:
            # print user if level is target
            print(current_user, end=' ')
            continue

        # otherwise, add user to visited
        visited.add(current_user)

        # search friends of current user
        for friend in g[current_user]:
            if friend not in visited:
                # add friends with updated levels to queue
                queue.append((friend, level + 1))
```

```python

g = {
  'Richard' : [],
  'Bob' : ['Richard', 'Rob', 'Pam'],
  'Rob' : [],
  'Pam' : ['Roger', 'Peter'],
  'Amy' : [],
  'Peter' : ['Amy'],
  'Roger' : ['Anna'],
  'Anna' : []
}
```

```python
FindFriends(g, 'Bob', 3)
```
