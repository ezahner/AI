**Task:** Write a DFS algorithm which when given a start node and a goal node, it returns SUCCESS if the goal is reachable and FAILURE otherwise. Your algorithm should test both graphs shown below and return SUCCESS for the first graph and FAILURE for the second one.

```python
def dfs(maze, current_node, visited=set()):
    # Goal node found
    if current_node == 50:
        return "Success! The path is found."

    # Check if the current node is already visited
    if current_node in visited:
        return "Failure. No path found."

    visited.add(current_node)

    # Explore neighbors recursively
    for neighbor in maze.get(current_node, []):
        if dfs(maze, neighbor, visited):
            return "Success! The path is found."

    return "Failure. No path found."
```
